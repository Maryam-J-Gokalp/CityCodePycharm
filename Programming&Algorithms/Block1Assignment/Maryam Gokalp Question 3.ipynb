```python
# # # Student Maryam Joana Henrique Da Silva Gokalp 
# Student Ref Number: 240064707

# QUESTIONS 3

#3)	Write a commented code to do the following:
#a)	Read the first line of a file called ‘numbers.txt’.
#b)	Record the numbers read from the file in a list of integers called my_numbers. Assume that the numbers are separated by commas.
# c)	Calculate the average of the numbers read
#d)	The code should then find the smallest even number in the list
#e)	The code should then print the following to the console. The string ‘The average is:’  followed by the average value, then on the next line print ‘The smallest even number is:’  followed by the smallest even number value.
# Program to read numbers from a file, process them, and display results

# Breakdown Questions with answers below 

# Task: Read numbers from a file, calculate the average, and find the smallest even number.

# Step 1: Open the file 'numbers.txt' and read the first line
try:
    # Use 'with' to safely open and automatically close the file
    with open('numbers.txt', 'r') as file:
        first_line = file.readline().strip()  # Read the first line and clean it up
except FileNotFoundError:
    # Handle the case where the file does not exist
    print("Error: The file 'numbers.txt' could not be found.")
    exit()  # Exit the program if the file is missing

# Step 2: Extract numbers from the line and convert them to integers
try:
    # Split the line into parts and convert each part to an integer
    numbers_list = [int(item) for item in first_line.split(',')]
except ValueError:
    # Handle the case where the data in the file is not properly formatted
    print("Error: The file contains invalid data. Ensure it has integers separated by commas.")
    exit()

# Step 3: Calculate the average of the numbers
if numbers_list:  # Ensure the list is not empty
    average_value = sum(numbers_list) / len(numbers_list)  # Calculate the average
else:
    # Handle the case where the list is empty
    print("Error: No numbers were found in the file.")
    exit()

# Step 4: Identify the smallest even number in the list
# Use a list comprehension to collect even numbers
even_numbers = [num for num in numbers_list if num % 2 == 0]
if even_numbers:  # Check if there are any even numbers
    smallest_even_number = min(even_numbers)  # Find the smallest even number
else:
    smallest_even_number = None  # Set to None if no even numbers are found

# Step 5: Output the results
# Print the calculated average
print(f"The average is: {average_value:.2f}")  # Format the average to two decimal places
# Print the smallest even number or a message if none exist
if smallest_even_number is not None:
    print(f"The smallest even number is: {smallest_even_number}")
else:
    print("There are no even numbers in the list.")

```
