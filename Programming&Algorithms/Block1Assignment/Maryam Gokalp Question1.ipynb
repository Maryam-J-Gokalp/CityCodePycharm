```python
# Student Maryam Joana Henrique Da Silva Gokalp 
# Student Ref Number: 240064707

# QUESTION 1

# Functionality: Filters numbers greater than 5 from a list and displays the result.
#1)	You are given the following uncommented code for function foo:
#def foo(a):
   # b = [j for j in a if j > 5]
	#print(b)

#a)	Provide a commented version of the code above. In a comment before the code explain the purpose of this piece of code. Then add clear comments in the rest of the code so that another user could understand it.
#b)	You now need to write a new version of the code. You now need to write the code with the same functionality but using a while loop instead of list comprehension. The code should contain clear comments.

#------------------------------------------------------------------------------
# This function filters numbers greater than 5 from the input list 'a'
# and prints the resulting list.
def foo(a):
    # Create a new list 'b' containing only elements from 'a' that are greater than 5
    b = [j for j in a if j > 5]
    
    # Print the filtered list of numbers greater than 5
    print(b)

"""

# NEW VERSION USING WHILE LOOP
# This function filters numbers greater than 5 from the input list 'a'
# and prints the resulting list using a while loop.

"""

def foo_using_while(a):
    # Initialize an empty list to store numbers greater than 5
    filtered_numbers = []
    
    # Initialize a counter to keep track of the current index
    index = 0

    # Iterate through the list until the end
    while index < len(a):
        # Check if the current element is greater than 5
        if a[index] > 5:
            # Add the element to the filtered list
            filtered_numbers.append(a[index])
        # Move to the next index
        index += 1

    # Print the filtered list of numbers greater than 5
    print(filtered_numbers)

# Example Usage
if __name__ == "__main__":
    # Define a sample list of numbers
    sample_list = [1, 6, 3, 8, 4, 10]

    # Call the original function
    print("Using the original function:")
    foo(sample_list)

    # Call the rewritten function
    print("Using the while loop function:")
    foo_using_while(sample_list)
```
