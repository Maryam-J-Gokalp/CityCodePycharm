```python
# # # Student Maryam Joana Henrique Da Silva Gokalp 
# Student Ref Number: 240064707

# QUESTION 5

""""
5)	Your manager now needs you to provide a more complex version of the code that you have in question 4. Now the code should:
a)	Check if the value entered was numeric. If it is a number, the message “The character entered was numeric” should be printed to the console and then the user should be asked again to enter a single letter.
b)	If a sting entered was multiple letters, then the code should take the first letter in a string and store it in variable my_char
c)	Then the program should ask the user to enter a number and store it in variable my_num. Check if the value entered is a number and if it is not ask the user to enter the number again.
d)	After my_char and my_num have been entered correctly the program should print to the console a string consisting of repeated my_char letters, the number of letters should be equal to my_num.
i)	Example: if my_num was 4 and my_char was y, the string printed should be ‘yyyy’.
"""


# Task: Process user inputs for a letter and a number with validations and perform specific actions

# Step 1: Get a valid single letter from the user
while True:
    user_input = input("Please enter a single letter: ").strip()  # Ask the user for input and remove extra spaces

    # Check if the input is numeric
    if user_input.isnumeric():
        print("The character entered was numeric.")  # Notify user if the input is a number
        continue  # Ask the user to input again

    # Check if the input contains multiple characters
    if len(user_input) > 1:
        my_char = user_input[0]  # Take the first character of the string
        print(f"Multiple characters were entered. Using the first letter: '{my_char}'.")
    else:
        my_char = user_input  # If input is a single character, assign it to my_char

    # Validate that the selected character is a letter
    if my_char.isalpha():  # Ensure it's a letter
        break  # Exit the loop once we have a valid letter
    else:
        print("The value entered was not a valid letter. Please try again.")

# Step 2: Get a valid number from the user
while True:
    num_input = input("Please enter a number: ").strip()  # Ask the user for a number

    if num_input.isnumeric():  # Check if the input is a valid number
        my_num = int(num_input)  # Convert the input to an integer
        break  # Exit the loop once we have a valid number
    else:
        print("The value entered was not a valid number. Please try again.")  # Notify user and retry

# Step 3: Print the character repeated my_num times
result = my_char * my_num  # Repeat the character my_num times
print(f"The resulting string is: '{result}'")  # Display the result
```
