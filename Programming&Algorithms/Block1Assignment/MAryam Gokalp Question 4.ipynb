```python
# # # Student Maryam Joana Henrique Da Silva Gokalp 
# Student Ref Number: 240064707

# QUESTION 4

#4)	Your manager asks to you write a code that:
#a)	Asks a user to enter a single letter and store it in a variable called my_char 
#b)	If my_char is not a single letter the program should print a message to the console saying that “the value entered was not a single letter”.
#c)	The program should then check if the letter is in uppercase. If so it should print to the console a message saying “An uppercase letter has been entered”. Otherwise it should print “A lowercase letter has been entered”.
#d)	Your code should be commented so that it is easy for another person to understand the code.


# Step 1: Prompt the user to enter a single letter
my_char = input("Please enter a single letter: ").strip()  # Take input and remove any extra spaces

# Step 2: Validate that the input is a single character
if len(my_char) != 1 or not my_char.isalpha():  # Check if the input is exactly one letter
    # Print an error message if the input is invalid
    print("The value entered was not a single letter.")
else:
    # Step 3: Determine if the letter is uppercase or lowercase
    if my_char.isupper():  # Check if the letter is uppercase
        print("An uppercase letter has been entered.")  # Notify about uppercase input
    else:
        print("A lowercase letter has been entered.")  # Notify about lowercase input
```
