```python
# # Student Maryam Joana Henrique Da Silva Gokalp 
# Student Ref Number: 240064707

"""
 QUESTION 2: 
Set Operations Program

a)	Write a program that takes 5 integers from user input and stores them in a set. Provide comments explaining the function of the code written.
b)	Write a function that takes two sets of integers as an argument and prints to the console a set of elements common to both sets and a set of elements that can be found in either or both sets
c)	Then write a function that takes a set of integers as an argument and prints the values one by one in descending order. 
d)	Provide comments explaining the function of the code written.
"""


# Part (a): Collect 5 unique integers from the user and store them in a set
def collect_user_set():
    """
    Collects 5 unique integers from the user and stores them in a set.
    Ensures only valid integers are added, and duplicate entries are avoided automatically.
    """
    user_numbers = set()  # Initialize an empty set
    print("Please enter 5 unique integers:")

    while len(user_numbers) < 5:  # Loop until exactly 5 unique integers are collected
        try:
            # Prompt the user for input and convert it to an integer
            number = int(input(f"Enter integer {len(user_numbers) + 1}: "))
            user_numbers.add(number)  # Add the integer to the set
        except ValueError:
            # Handle cases where the input is not a valid integer
            print("Invalid input. Please enter an integer.")

    return user_numbers  # Return the set of user-provided integers


# Part (b): Analyze two sets for common and union elements
def analyze_two_sets(set_a, set_b):
    """
    Takes two sets of integers as input and prints:
    - The elements common to both sets (intersection).
    - All elements present in either or both sets (union).
    """
    common = set_a.intersection(set_b)  # Find common elements between the sets
    union = set_a.union(set_b)  # Combine elements from both sets

    print("Common elements between the two sets:", common)
    print("Union of both sets:", union)


# Part (c): Print elements of a set in descending order
def print_set_in_descending_order(input_set):
    """
    Prints all elements of a set in descending order.
    The elements are sorted and displayed one by one.
    """
    sorted_set = sorted(input_set, reverse=True)  # Sort the set in descending order
    print("Set elements in descending order:")
    for value in sorted_set:  # Loop through the sorted values and print each
        print(value)


# Example Usage
if __name__ == "__main__":
    # --- Part (a): Collect Unique Integers from the User ---
    print("\n--- Part (a): Collect Unique Integers ---")
    user_set = collect_user_set()
    print("Set created from user input:", user_set)

    # --- Part (b): Analyze Two Example Sets ---
    print("\n--- Part (b): Analyze Two Sets ---")
    set_one = {10, 20, 30, 40, 50}  # Example Set 1
    set_two = {30, 40, 50, 60, 70}  # Example Set 2
    print("Set 1:", set_one)
    print("Set 2:", set_two)
    analyze_two_sets(set_one, set_two)

    # --- Part (c): Print User Set in Descending Order ---
    print("\n--- Part (c): Print Set in Descending Order ---")
    print_set_in_descending_order(user_set)

```
